-- Criando a estrutura das tabelas para armazenar os dados

-- Tabela para os dados financeiros das operadoras
CREATE TABLE demonstracoes_contabeis (
    id SERIAL PRIMARY KEY,
    trimestre VARCHAR(10),
    ano INT,
    codigo_operadora VARCHAR(20),
    nome_operadora VARCHAR(255),
    despesas_eventos_saude NUMERIC(15,2)
);

-- Tabela para os dados cadastrais das operadoras
CREATE TABLE operadoras_ativas (
    id SERIAL PRIMARY KEY,
    codigo_operadora VARCHAR(20) UNIQUE,
    nome_operadora VARCHAR(255),
    cnpj VARCHAR(18),
    modalidade VARCHAR(100)
);

-- Importação dos arquivos CSV (Substitua 'caminho_arquivo' pelo caminho real)
-- No PostgreSQL, use a função COPY:
COPY demonstracoes_contabeis(trimestre, ano, codigo_operadora, nome_operadora, despesas_eventos_saude)
FROM 'caminho_arquivo/1T2023.csv'
DELIMITER ';' CSV HEADER ENCODING 'UTF8';

COPY demonstracoes_contabeis(trimestre, ano, codigo_operadora, nome_operadora, despesas_eventos_saude)
FROM 'caminho_arquivo/2T2024.csv'
DELIMITER ';' CSV HEADER ENCODING 'UTF8';

COPY operadoras_ativas(codigo_operadora, nome_operadora, cnpj, modalidade)
FROM 'caminho_arquivo/Relatorio_cadop.csv'
DELIMITER ';' CSV HEADER ENCODING 'UTF8';

-- No MySQL, use LOAD DATA INFILE:
-- Certifique-se de que o MySQL permite LOAD DATA LOCAL INFILE
-- (pode ser necessário ativá-lo com --local-infile=1)
LOAD DATA LOCAL INFILE 'caminho_arquivo/1T2023.csv'
INTO TABLE demonstracoes_contabeis
FIELDS TERMINATED BY ';'
LINES TERMINATED BY '\n'
IGNORE 1 LINES;

LOAD DATA LOCAL INFILE 'caminho_arquivo/2T2024.csv'
INTO TABLE demonstracoes_contabeis
FIELDS TERMINATED BY ';'
LINES TERMINATED BY '\n'
IGNORE 1 LINES;

LOAD DATA LOCAL INFILE 'caminho_arquivo/Relatorio_cadop.csv'
INTO TABLE operadoras_ativas
FIELDS TERMINATED BY ';'
LINES TERMINATED BY '\n'
IGNORE 1 LINES;

-- Consulta para as 10 operadoras com maiores despesas no último trimestre
SELECT nome_operadora, SUM(despesas_eventos_saude) AS total_despesas
FROM demonstracoes_contabeis
WHERE ano = 2024 AND trimestre = '2T'
GROUP BY nome_operadora
ORDER BY total_despesas DESC
LIMIT 10;

-- Consulta para as 10 operadoras com maiores despesas no último ano
SELECT nome_operadora, SUM(despesas_eventos_saude) AS total_despesas
FROM demonstracoes_contabeis
WHERE ano = 2023
GROUP BY nome_operadora
ORDER BY total_despesas DESC
LIMIT 10;
